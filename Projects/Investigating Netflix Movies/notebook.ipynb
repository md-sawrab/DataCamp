```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt

# Read in the Netflix CSV as a DataFrame
netflix_df = pd.read_csv("netflix_data.csv")
```

```python
netflix_df.head(10)
```

```python
netflix_subset = netflix_df[netflix_df['type'] == 'Movie']

netflix_subset
```

```python
netflix_filter = netflix_subset[(netflix_subset['release_year'] >= 1990) & (netflix_subset['release_year'] < 2000)]

netflix_filter
```

```python
plt.hist(netflix_filter["duration"])
plt.title('Distribution of Movie Durations in the 1990s')
plt.xlabel('Duration (minutes)')
plt.ylabel('Number of Movies')

plt.show()
```

```python
action_movies = netflix_filter[netflix_filter["genre"] == "Action"]

action_movies
```

```python
short_movie_count = 0  # Initialize the counter

for label, row in action_movies.iterrows():
    if row["duration"] < 90:
        short_movie_count += 1

print(short_movie_count)
```
