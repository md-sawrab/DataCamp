#### Filtering Arrays

there are two main ways to numpy filtering

- Mask and fancy indexing
- np.where()

```python
# mask

one_to_five = np.arange(1,5)

mask = one_to_five % 2 == 0
mask
```

```python
# fancy indexing

one_to_five[mask]
```

```python
# 2D fancy indexing

class_ids_and_size = np.array([[1,22],[2,21],[3,27],[4,26]])

even_no_stu = class_ids_and_size[:,1] % 2 == 0

class_ids_and_size[:,0][even_no_stu]
```

```python
# np.where(): return the array of indices

class_ids_and_size

np.where(class_ids_and_size[:, 1] % 2 == 0) # return the tupele of arrays
```

```python
# A tuple of indicies

row_ind, col_ind = np.where(sudoku_game == 0)

row_ind, col_ind
```

```python
# find and repale

np.where(sudoku_game == 0, " ", sudoku_game)
```

#### Adding and Removing

```python
# concatenating rows 

class_ids_and_size = np.array([[1,22],[2,21],[3,27],[4,26]])
new_class = np.array([[5,23]])

np.concatenate((class_ids_and_size, new_class))
```

```python
# concatenating columns

class_ids_and_size = np.array([[1,22],[2,21],[3,27],[4,26]])
new_class = np.array([[5,22],[6,21],[7,27],[8,26]]) # must have same shape

concatenate = np.concatenate((class_ids_and_size, new_class), axis = 1)
concatenate
```

Two array must be same number of dimesion

```python
# delete rows

np.delete(class_ids_and_size, 1, axis = 0)
```

```python
# cloumn delete

np.delete(concatenate, 1 , axis = 1)
```

```python
# monthly sales example

monthly_sales
```

```python
# which totals the monthly sales across industries for each month

monthly_industry_sales = monthly_sales.sum(axis = 1, keepdims = True)

monthly_sales_with_total = np.concatenate((monthly_sales,monthly_industry_sales), axis = 1)

monthly_sales_with_total
```

```python
# plot

avg_monthly_sales = monthly_sales.mean(axis=1)
print(avg_monthly_sales)

plt.plot(np.arange(1,13), avg_monthly_sales, label="Average sales across industries")

plt.plot(np.arange(1,13), monthly_sales[:,2], label="Department store sales")
plt.legend()
plt.show()
```

```python
cumulative_monthly_industry_sales = monthly_sales.cumsum(axis = 0)

plt.plot(np.arange(1, 13),monthly_sales[:,0].cumsum(axis = 0) , label="Liquor Stores")
plt.plot(np.arange(1, 13), monthly_sales[:,1].cumsum(axis = 0), label="Restaurants")
plt.plot(np.arange(1, 13), monthly_sales[:,2].cumsum(axis = 0), label="Department stores")
plt.legend()
plt.show()
```

```python
# projected Sales

monthly_industry_multipliers = np.array([[0.98, 1.02, 1.  ],
       [1.00, 1.01, 0.97],
       [1.06, 1.03, 0.98],
       [1.08, 1.01, 0.98],
       [1.08, 0.98, 0.98],
       [1.1 , 0.99, 0.99],
       [1.12, 1.01, 1.  ],
       [1.1 , 1.02, 1.  ],
       [1.11, 1.01, 1.01],
       [1.08, 0.99, 0.97],
       [1.09, 1.  , 1.02],
       [1.13, 1.03, 1.02]]) # monthly_industry_multipliers[0, 0] of 0.98 means that the liquor store industry is projected to have 98% of this January's sales in January of next year.

projected_monthly_sales = monthly_sales * monthly_industry_multipliers

plt.plot(np.arange(1, 13), monthly_sales[:, 0], label="Current liquor store sales")
plt.plot(np.arange(1,13), projected_monthly_sales[:,0], label="Projected liquor store sales")
plt.legend()
plt.show()
```

### Array Transformation

#### Saving and loading arrays

Save arrays in many formate

- .CSV

- .pkl

```python
with open("datasets/rgb_array.npy", "rb") as f: # the first argument is name of the file to open and second is mode to ope the file in.
    rgb_array = np.load(f)
plt.imshow(rgb_array)
plt.show()
```

```python
# update and save

darker_rgb_array = rgb_array * 0.5

# Convert darker_rgb_array into an array of integers
darker_rgb_int_array = darker_rgb_array.astype(np.int8)
plt.imshow(darker_rgb_int_array)
plt.show()

# Save darker_rgb_int_array to an .npy file called darker_monet.npy
with open("darker_monet.npy", "wb") as f:
    np.save(f, darker_rgb_int_array)
```
